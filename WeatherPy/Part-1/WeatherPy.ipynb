```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import sys
import csv 


# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "data/cities.csv"

#sys.path.append("../../../")

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
print (query_url)
```

```python
response=requests.get(query_url+"new york")
print(response.json())
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#data=response.json()
```

```python
#temperature=data['main']['temp']
#temperature
```

```python
#loop through all cities 
weather_data=[]
i=0
for city in cities:
     
    data=requests.get(query_url + city).json()
    i=i+1
#use the try and except block
    try:
        city_name=data["name"]
        temp_max=data["main"]['temp_max']
        humudity=data["main"]['temp_max']
        speed= data["wind"]["speed"]
        cloudyness=data['clouds']['all']
        lattitude=data['coord']['lat']
        country=data['sys']['country']
        date=data['dt']
        longitude=data['coord']['lon']
        weather_data.append([city_name, temp_max,humudity,speed,cloudyness,lattitude,country,date,longitude])
        print(f"counter {i} || {city}")
        time.sleep(.1)
    except:
        print('City not Found')
```

```python
#display the DataFrame 
data_ds=pd.DataFrame(weather_data,columns=["City","Temperature Max", 'Humidity','Speed','Cloudiness',
                                           "Lattitude", "Country","Date","Longitude"])
data_ds.head()
```

```python
# Reset index of dataframe
data_ds = data_ds.reset_index(drop=True)

# Convert unix timestamps to dates
data_ds['Date'] = pd.to_datetime(data_ds['Date'])
data_ds['Date'] = data_ds['Date'].map(lambda x: x.strftime('%m/%d/%y'))
data_ds.head()
```

```python
#city data to a csv
data_ds.to_csv(output_data_file, index=True)
```
